```python
from getdist import loadMCSamples, plots
```

```python
samples_george_mbac = loadMCSamples('../geroge_act_plik_common_poisson/outputs/chains/george_act', settings={'ignore_rows':0.3})
samples_george_mbac_plik = loadMCSamples('../geroge_act_plik_common_poisson/outputs/chains/plik_george_act', settings={'ignore_rows':0.3})
samples_reichardt_mbac = loadMCSamples('../reichardt_act_plik_common_poisson/outputs/chains/reichardt_act', settings={'ignore_rows':0.3})
samples_reichardt_mbac_plik = loadMCSamples('../reichardt_act_plik_common_poisson/outputs/chains/plik_reichardt_act', settings={'ignore_rows':0.3})
```

```python
params_common = ['a_tSZ', 'a_kSZ', 'a_CIB', 'xi', 'beta_c', 'a_gtt_as', 'a_gtt_ae', 'a_gtt_spt', 'a_r_a', 'a_r_s', 'a_CIBp']
g = plots.get_subplot_plotter(width_inch=20)
g.plots_1d([samples_george_mbac, samples_george_mbac_plik],
           params_common, nx=4, legend_labels=[ "George+MBAC (DR6 modelling)", "George+MBAC+plik (DR6 modelling)", "Reichardt+MBAC+plik (DR6 modelling)", "Reichardt+MBAC+plik (DR6 modelling)"])
```
