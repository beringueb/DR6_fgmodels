## Planck

```python
samples_plik = loadMCSamples('../highL_plik/outputs/chains/plik', settings={'ignore_rows':0.3})
samples_planck_commonSZ = loadMCSamples('../highL_plik_commonSZ/outputs/chains/plik', settings={'ignore_rows':0.3})
samples_planck_commonSZCIB = loadMCSamples('../highL_plik_commonSZCIB/outputs/chains/plik', settings={'ignore_rows':0.3})
samples_planck_commonSZCIB_fixedTc = loadMCSamples('../highL_plik_commonSZCIB_fixedTc/outputs/chains/plik', settings={'ignore_rows':0.3})
```

```python
params_common = ['a_tSZ', 'a_kSZ', 'a_CIB', 'xi', 'T_dc', 'beta_c']
```

```python
g = plots.get_subplot_plotter(width_inch=20)
g.plots_1d([ samples_planck_commonSZCIB, samples_plik, samples_planck_commonSZ, samples_planck_commonSZCIB_fixedTc],
           params_common, nx=3, legend_labels=["Common SZ/CIB templates, MBB", "plik", "Common SZ","Common SZ/CIB templates, MBB fixed T_eff" ])
```
