```python
from getdist import loadMCSamples, plots
```

```python
samples_MBAC_90_base = loadMCSamples("../DR4_MBAC_base/outputs/chains/MBAC_90", settings={'ignore_rows':0.3})
samples_DR4_220_base = loadMCSamples("../DR4_MBAC_base/outputs/chains/DR4_220", settings={'ignore_rows':0.3})
samples_DR4_base = loadMCSamples("../DR4_MBAC_base/outputs/chains/DR4_priorcibradio", settings={'ignore_rows':0.3})
samples_MBAC_base = loadMCSamples("../DR4_MBAC_base/outputs/chains/MBAC", settings={'ignore_rows':0.3})
samples_DR4_220_base_prior = loadMCSamples("../DR4_MBAC_base/outputs/chains/DR4_220_priorcibradio", settings={'ignore_rows':0.3})

samples_DR4_prior = loadMCSamples("../DR4_MBAC_base/outputs/chains/DR4_priorsamp", settings={'ignore_rows':0.3})
```

```python
params = ['a_tSZ', 'a_kSZ','xi','a_CIBp','a_r_a','a_r_aw',
          'a_CIB','beta_c', 'a_gtt_ad', 'a_gtt_aw',
         ]

g = plots.get_subplot_plotter(width_inch=20)
g.plots_1d([samples_DR4_220_base_prior, samples_DR4_base, samples_MBAC_base],
           params, nx=6, legend_labels=["DR4 + MBAC 220x220", "DR4 only",  "MBAC only"])
```
