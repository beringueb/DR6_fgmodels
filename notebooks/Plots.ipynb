```python
from getdist import loadMCSamples, plots
```

```python
samples_DR4_220_base = loadMCSamples("../DR4_MBAC_base/outputs/chains/DR4_220", settings={'ignore_rows':0.3})
samples_DR4_220_base_prior = loadMCSamples("../DR4_MBAC_base/outputs/chains/DR4_220_priorcibradio", settings={'ignore_rows':0.3})
samples_DR4_220_base_prior_plik = loadMCSamples("../DR4_MBAC_base/outputs/chains/DR4_220_plik_prior", settings={'ignore_rows':0.3})


samples_DR4_220_betacp = loadMCSamples("../DR4_MBAC_betac-betap/outputs/chains/DR4_220", settings={'ignore_rows':0.3})
samples_DR4_220_betacp_prior = loadMCSamples("../DR4_MBAC_betac-betap/outputs/chains/DR4_220_priorcibradio", settings={'ignore_rows':0.3})
samples_DR4_220_betacp_prior_plik = loadMCSamples("../DR4_MBAC_betac-betap/outputs/chains/DR4_220_plik_prior", settings={'ignore_rows':0.3})

samples_DR4_220_betacpr = loadMCSamples("../DR4_MBAC_betac-betap-betar/outputs/chains/DR4_220", settings={'ignore_rows':0.3})
samples_DR4_220_betacpr_prior = loadMCSamples("../DR4_MBAC_betac-betap-betar/outputs/chains/DR4_220_priorcibradio", settings={'ignore_rows':0.3})
samples_DR4_220_betacpr_prior_plik = loadMCSamples("../DR4_MBAC_betac-betap-betar/outputs/chains/DR4_220_plik_prior", settings={'ignore_rows':0.3})


samples_DR4_220_betar = loadMCSamples("../DR4_MBAC_betar/outputs/chains/DR4_220", settings={'ignore_rows':0.3})
samples_DR4_220_betar_prior = loadMCSamples("../DR4_MBAC_betar/outputs/chains/DR4_220_priorcibradio", settings={'ignore_rows':0.3})
samples_DR4_220_betar_prior_plik = loadMCSamples("../DR4_MBAC_betar/outputs/chains/DR4_220_plik_prior", settings={'ignore_rows':0.3})

```

```python
params = ['a_tSZ', 'a_kSZ','xi','a_CIBp','a_r_a','a_r_aw',
          'a_CIB','beta_c','beta_p','beta_r','a_gtt_ad', 'a_gtt_aw',
         ]

g = plots.get_subplot_plotter(width_inch=20)
g.plots_1d([samples_DR4_220_betacpr, samples_DR4_220_betacp, samples_DR4_220_betar, samples_DR4_220_base],
           params, nx=6, legend_labels=["DR4 + MBAC 220x220 beta_c != beta_p, beta_r free", 
                                        "DR4 + MBAC 220x220 beta_c != beta_p, beta_r fixed",
                                        "DR4 + MBAC 220x220 beta_c = beta_p, beta_r free", 
                                        "DR4 + MBAC 220x220 beta_c = beta_p, beta_r fixed"])
```
