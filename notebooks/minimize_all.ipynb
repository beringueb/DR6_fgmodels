```python
def run2min_file(fname):
    with open(fname, 'r') as f:
        lines = f.readlines()
        lines_out = lines.copy()
        for i, line in enumerate(lines):
            if line[:6]=='resume':
                lines_out[i] = '\n'
            if line[:7] =='  mcmc:':
                lines_out[i]='  minimize:\n'
    with open(fname, 'w') as f:
        f.writelines(lines_out)
```
