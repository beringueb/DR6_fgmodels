```python
from getdist import loadMCSamples, plots
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
samples_arp_dr6 = loadMCSamples("../DR6_modelling/outputs/chains_2/MBAC_reichardt_plik_fixedcosmo", settings={'ignore_rows':0.3})
samples_arp_common = loadMCSamples("../common_templates/outputs/chains_2/MBAC_reichardt_plik_fixedcosmo", settings={'ignore_rows':0.3})

samples_agp_dr6 = loadMCSamples("../DR6_modelling/outputs/chains_2/MBAC_george_plik_fixedcosmo", settings={'ignore_rows':0.3})
samples_agp_common = loadMCSamples("../common_templates/outputs/chains_2/MBAC_george_plik_fixedcosmo", settings={'ignore_rows':0.3})
```

```python
params = ['a_tSZ', 'a_kSZ','a_CIB','xi','beta_c', 
          'a_gtt_as','a_gtt_ae', 'a_gtt_spt', 'a_r_a', 'a_r_s', 
          'a_CIBp','beta_p', 'cal_acts_148','cal_acts_220',
          'cal_acte_148','cal_acte_220','cal_sptr_90','cal_sptr_150','cal_sptr_220']
```

```python
g = plots.get_subplot_plotter(width_inch=20)
g.plots_1d([samples_arp_dr6, samples_arp_common, samples_agp_dr6, samples_agp_common],
           params, nx=5, legend_labels=["Reichardt+MBAC+plik (DR6 model)", "Reichardt+MBAC+plik (common_templates)",
                                       "George+MBAC+plik (DR6 model)", "George+MBAC+plik (common_templates)"])

# plt.savefig("../figures/Geroge_reichardt_MBAC_plik.png")
```
