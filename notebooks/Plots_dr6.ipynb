```python
from getdist import loadMCSamples, plots
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
samples_arp_dr6 = loadMCSamples("../DR6_modelling/outputs/chains_2/MBAC_reichardt_plik_fixedcosmo", settings={'ignore_rows':0.3})
samples_arp_common = loadMCSamples("../common_templates/outputs/chains_2/MBAC_reichardt_plik_fixedcosmo", settings={'ignore_rows':0.3})

samples_arp_dr6_beta_r = loadMCSamples("../DR6_modelling/outputs/chains_2/MBAC_reichardt_plik_betar_fixedcosmo", settings={'ignore_rows':0.3})
samples_arp_common_Tdp97 = loadMCSamples("../common_templates/outputs/chains_2/MBAC_reichardt_plik_Tdp97_fixedcosmo", settings={'ignore_rows':0.3})

samples_arp_original = loadMCSamples("../original/outputs/chains_2/MBAC_reichardt_plik_fixedcosmo", ranges ={'a_CIB':[0,10], 'a_CIBp':[0,20]}, settings={'ignore_rows':0.3})
```

```python
params = ['a_tSZ', 'a_kSZ','a_CIB','xi','beta_c', 
          'a_gtt_as','a_gtt_ae', 'a_gtt_spt', 'a_r_a', 'a_r_s', 
          'a_CIBp','beta_p', 'cal_acts_148','cal_acts_220',
          'cal_acte_148','cal_acte_220','cal_sptr_90','cal_sptr_150','cal_sptr_220']
```

```python
g = plots.get_subplot_plotter(width_inch=20)
g.plots_1d([samples_arp_dr6, samples_arp_common],
           params, nx=5, legend_labels=["Reichardt+MBAC+plik (DR6 model)", "Reichardt+MBAC+plik (common_templates)"])

plt.savefig("../figures/1509reichardt_MBAC_plik.png")
```

```python
g = plots.get_subplot_plotter(width_inch=20)
g.plots_1d([samples_arp_dr6_beta_r, samples_arp_dr6, samples_arp_common, samples_arp_common_Tdp97],
           params + ['beta_r'], nx=5, legend_labels=["Reichardt+MBAC+plik (DR6 model + varying beta_r)", "Reichardt+MBAC+plik (DR6 model)", "Reichardt+MBAC+plik (common templates Tdp=25K)", "Reichardt+MBAC+plik (common templates Tdp=9.7K)"])

plt.savefig("../figures/1509reichardt_MBAC_plik_poisson.png")
```

```python
g = plots.get_subplot_plotter(width_inch=20)
g.plots_1d([samples_arp_dr6_beta_r, samples_arp_dr6, samples_arp_common_Tdp97],
           params + ['beta_r'], nx=5, legend_labels=["Reichardt+MBAC+plik (DR6 model + varying beta_r)", "Reichardt+MBAC+plik (DR6 model)","Reichardt+MBAC+plik (common templates Tdp=9.7K)"])

```
