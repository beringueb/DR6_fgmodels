```python
from getdist import loadMCSamples, plots
import matplotlib.pyplot as plt
```

```python
samples_agp_dr6 = loadMCSamples("../DR6_modelling/outputs/chains/MBAC_george_plik_fixedcosmo", settings={'ignore_rows':0.3})
samples_ag_dr6 = loadMCSamples("../DR6_modelling/outputs/chains/MBAC_george_fixedcosmo", settings={'ignore_rows':0.3})

samples_arp_dr6 = loadMCSamples("../DR6_modelling/outputs/chains/MBAC_reichardt_plik_fixedcosmo", settings={'ignore_rows':0.3})
samples_ar_dr6 = loadMCSamples("../DR6_modelling/outputs/chains/MBAC_reichardt_fixedcosmo", settings={'ignore_rows':0.3})

samples_arp_common = loadMCSamples("../common_templates/outputs/chains/MBAC_reichardt_plik_fixedcosmo", settings={'ignore_rows':0.3})
samples_agp_common = loadMCSamples("../common_templates/outputs/chains/MBAC_george_plik_fixedcosmo", settings={'ignore_rows':0.3})
```

```python
params_common = ['a_tSZ', 'a_kSZ','a_CIB','xi','beta_c']
params_highL = ['a_gtt_as','a_gtt_ae', 'a_gtt_spt', 'a_r_a', 'a_r_s', 'a_CIBp','beta_p', 'beta_r', 'cal_acts_148',
                'cal_acts_220','cal_acte_148','cal_acte_220']
```

```python
g = plots.get_subplot_plotter(width_inch=20)
g.plots_1d([samples_agp_dr6, samples_arp_dr6],
           params_common, nx=3, legend_labels=["George+MBAC+plik (DR6 model)", "Reichardt+MBAC+plik (DR6 model)"])

plt.savefig("../figures/Geroge_reichardt_MBAC_plik.png")
```

```python
g = plots.get_subplot_plotter(width_inch=20)
g.plots_1d([samples_agp_dr6, samples_arp_dr6, samples_agp_common, samples_arp_common],
           params_common, nx=3, legend_labels=["George+MBAC+plik (DR6 model)", "Reichardt+MBAC+plik (DR6 model)","George+MBAC+plik (common templates)", "Reichardt+MBAC+plik (common templates)"])

plt.savefig("../figures/Geroge_reichardt_MBAC_plik_common.png")
```

```python
g = plots.get_subplot_plotter(width_inch=20)
g.plots_1d([samples_ar_dr6, samples_arp_dr6, samples_ag_dr6, samples_agp_dr6],
           params_highL, nx=6, legend_labels=["Reichardt+MBAC (DR6 model)", "Reichardt+MBAC+plik (DR6 model)", "George+MBAC (DR6 model)", "George+MBAC+plik (DR6 model)"])
```

```python
g = plots.get_subplot_plotter(width_inch=20)
g.plots_1d([samples_ar_common, samples_arp_common, samples_ar_dr6, samples_arp_dr6],
           params_common, nx=3, legend_labels=[ "Reichardt+MBAC (Commn templates)", "Reichardt+MBAC+plik (Commn templates)","Reichardt+MBAC (DR6 model)", "Reichardt+MBAC+plik (DR6 model)"])
```
