```python
from getdist import loadMCSamples, plots
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
samples_ar_original = loadMCSamples("../original/outputs/chains_2/MBAC_reichardt_fixedcosmo", settings={'ignore_rows':0.3})
samples_ar_common = loadMCSamples("../common_templates/outputs/chains_2/MBAC_reichardt_Tdp97_fixedcosmo", settings={'ignore_rows':0.3})
samples_ar_dr6 = loadMCSamples("../DR6_modelling/outputs/chains_2/MBAC_reichardt_betar_fixedcosmo", settings={'ignore_rows':0.3})

samples_r_original = loadMCSamples("../original/outputs/chains_2/reichardt_150_fixedcosmo", settings={'ignore_rows':0.3})
samples_r_common = loadMCSamples("../common_templates/outputs/chains_2/reichardt_Tdp97_fixedcosmo", settings={'ignore_rows':0.3})
samples_r_dr6 = loadMCSamples("../DR6_modelling/outputs/chains_2/reichardt_betar_fixedcosmo", settings={'ignore_rows':0.3})

samples_ar_dr6_prior = loadMCSamples("../DR6_modelling/outputs/chains_2/MBAC_reichardt_prior_radio_fixedcosmo", settings={'ignore_rows':0.3})
params = ['a_tSZ', 'a_kSZ','a_CIB','a_gtt_spt', 'a_gtt_ae', 'a_gtt_as',
          'a_r_a', 'a_r_s', 'a_CIBp','beta_c','beta_p','beta_r', 
          'cal_sptr_90','cal_sptr_150','cal_sptr_220','cal_acte_148',
          'cal_acte_220','cal_acts_148','cal_acts_220' ]

g = plots.get_subplot_plotter(width_inch=20)
g.plots_1d([samples_ar_dr6, samples_ar_common, samples_ar_dr6_prior],
           params, nx=4, legend_labels=["ar DR6","ar common", "ar DR6 prior"])
```

```python
samples_ar_dr6_alpha = loadMCSamples("../DR6_modelling_alpha/outputs/chains_2/MBAC_reichardt_alpha_fixedcosmo", settings={'ignore_rows':0.3})
samples_ar_dr6 = loadMCSamples("../DR6_modelling/outputs/chains_2/MBAC_reichardt_betar_fixedcosmo", settings={'ignore_rows':0.3})

params = ['a_tSZ', 'a_kSZ','a_CIB','a_gtt_spt', 'a_gtt_ae', 'a_gtt_as',
          'a_r_a', 'a_r_s','beta_c','beta_r', 'alpha_CIB',
          'cal_sptr_90','cal_sptr_150','cal_sptr_220','cal_acte_148',
          'cal_acte_220','cal_acts_148','cal_acts_220' ]

g = plots.get_subplot_plotter(width_inch=20)
g.plots_1d([samples_ar_dr6_alpha, samples_ar_dr6],
           params, nx=4, legend_labels=["ar DR6 alpha", "ar DR6"])
```

```python
samples_ar_dr6_ampbetap = loadMCSamples("../DR6_modelling_amp_betap/outputs/chains_2/MBAC_reichardt_ampbetap_fixedcosmo", settings={'ignore_rows':0.3})
samples_ar_dr6_amp= loadMCSamples("../DR6_modelling_amp/outputs/chains_2/MBAC_reichardt_poissonamp_fixedcosmo", settings={'ignore_rows':0.3})
samples_ar_dr6_betap = loadMCSamples("../DR6_modelling_betap/outputs/chains_2/MBAC_reichardt_betap_fixedcosmo", settings={'ignore_rows':0.3})
samples_ar_dr6 = loadMCSamples("../DR6_modelling/outputs/chains_2/MBAC_reichardt_betar_fixedcosmo", settings={'ignore_rows':0.3})
samples_ar_dr6.addDerived(samples_ar_dr6["a_CIBp"], "a_CIBp_a")
samples_ar_dr6.addDerived(samples_ar_dr6["a_CIBp"], "a_CIBp_s")
samples_ar_dr6.addDerived(samples_ar_dr6["beta_p"], "beta_p_a")
samples_ar_dr6.addDerived(samples_ar_dr6["beta_p"], "beta_p_s")
params = ['a_tSZ', 'a_kSZ','a_CIB','a_gtt_spt', 'a_gtt_ae', 'a_gtt_as',
          'a_r_a', 'a_r_s', 'a_CIBp_a', 'a_CIBp_s','beta_c','beta_p_a', 'beta_p_s', 'beta_r',
          'cal_sptr_90','cal_sptr_150','cal_sptr_220','cal_acte_148',
          'cal_acte_220','cal_acts_148','cal_acts_220' ]

g = plots.get_subplot_plotter(width_inch=20)
g.plots_1d([samples_ar_dr6_ampbetap, samples_ar_dr6_amp, samples_ar_dr6_betap, samples_ar_dr6],
           params, nx=4, legend_labels=["beta free, amp free","beta same, amp free", 
                                        "beta free, amp same", "beta same, amp same"])
```
