```python
from getdist import loadMCSamples, plots
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
samples_r_base = loadMCSamples("../DR6_modelling_base/outputs/chains/reichardt", settings={'ignore_rows':0.3})
samples_a_base = loadMCSamples("../DR6_modelling_base/outputs/chains/MBAC_prior", settings={'ignore_rows':0.3})
samples_ar_base = loadMCSamples("../DR6_modelling_base/outputs/chains/MBAC_reichardt", settings={'ignore_rows':0.3})
samples_prior = loadMCSamples("../DR6_modelling_base/outputs/chains/MBAC_reichardt_one", settings={'ignore_rows':0.3})

params = ['a_tSZ', 'a_kSZ','a_CIB','a_gtt_spt', 'a_gtt_ae', 'a_gtt_as',
          'a_r_s', 'a_r_a', 'a_CIBp','beta_c', 'cal_acts_148', 'cal_acts_220', 
          'cal_acte_148', 'cal_acte_220','cal_sptr_90','cal_sptr_150','cal_sptr_220','xi']

g = plots.get_subplot_plotter(width_inch=20)
g.plots_1d([samples_ar_base, samples_r_base, samples_a_base, samples_prior],
           params, nx=4, legend_labels=["MBAC+ Reichardt (base)",
                                        "Reichardt (base)",
                                        "MBAC (base)"])


```

```python
g = plots.get_subplot_plotter(width_inch=20)
params_triangle = ['a_CIB', 'a_r_s', 'a_r_a', 'a_CIBp','beta_c', 'cal_acts_148', 'cal_acts_220',
                   'cal_acte_148', 'cal_acte_220','cal_sptr_90','cal_sptr_150','cal_sptr_220']

g.triangle_plot([samples_ar_base, samples_r_base, samples_a_base], params_triangle, 
    filled=True)
```
