```python
from getdist import loadMCSamples, plots
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
samples_r_original = loadMCSamples("../original/outputs/chains_2/reichardt_150_fixedcosmo", settings={'ignore_rows':0.3})
samples_ar_original = loadMCSamples("../original/outputs/chains_2/MBAC_reichardt_fixedcosmo", settings={'ignore_rows':0.3})
samples_arp_original = loadMCSamples("../original/outputs/chains_2/MBAC_reichardt_plik_fixedcosmo", settings={'ignore_rows':0.3})

params = ['a_tSZ', 'a_kSZ','a_CIB','a_CIB_1H', 'a_CIB_2H', 'xi','beta_c', 
          'a_gtt_spt', 'a_r_s', 'a_CIBp','beta_p',
          'cal_sptr_90','cal_sptr_150','cal_sptr_220']

g = plots.get_subplot_plotter(width_inch=20)
g.plots_1d([samples_r_original, samples_ar_original, samples_arp_original],
           params, nx=5, legend_labels=["Reichardt", "MBAC+Reichardt", "MBAC+Reichardt+plik"])
```
