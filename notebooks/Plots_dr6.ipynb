```python
from getdist import loadMCSamples, plots
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
samples_r_base = loadMCSamples("../DR6_modelling_base/outputs/chains/reichardt", settings={'ignore_rows':0.3})
samples_r_betar = loadMCSamples("../DR6_modelling_betar/outputs/chains/reichardt", settings={'ignore_rows':0.3})
samples_r_betac_betap = loadMCSamples("../DR6_modelling_betac-betap/outputs/chains/reichardt", settings={'ignore_rows':0.3})
samples_r_betac_betap_betar = loadMCSamples("../DR6_modelling_betac-betap-betar/outputs/chains/reichardt", settings={'ignore_rows':0.3})

params = ['a_tSZ', 'a_kSZ','a_CIB','a_gtt_spt',
          'a_r_s', 'a_CIBp','beta_c','beta_p','beta_r', 
          'cal_sptr_90','cal_sptr_150','cal_sptr_220']

g = plots.get_subplot_plotter(width_inch=20)
g.plots_1d([samples_r_betac_betap_betar, samples_r_base, samples_r_betac_betap, samples_r_betar],
           params, nx=4, legend_labels=["Reichardt DR6 (beta_c, beta_r, beta_p free)",
                                        "Reichardt DR6 (beta_c = beta_p, beta_r=-2.5)",
                                        "Reichardt DR6 (beta_c, beta_r, free, beta_r=-2.5)",
                                        "Reichardt DR6 (beta_c=beta_r, beta_p free)"])
```
